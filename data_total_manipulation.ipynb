```python
import pandas as pd
import matplotlib.pyplot as plt

df_total = pd.read_csv("data\population_total.csv", quotechar='"')
```

```python
df_total = df_total.drop(columns=["Country Code", "Indicator Name", "Indicator Code", "Unnamed: 68"])
df_total.head()
```

```python
df_total.set_index('Country Name', inplace=True)
df_total = df_total.T
```

```python
df_total.index.name = "Years"
```

```python
df_total.isnull().sum().sum()
```

```python
df_total = df_total.drop(columns=["World"])
```

```python
df_total
```

```python
plt.figure(figsize=(10, 5))
for column in df_total.columns[:10]:
    plt.plot(df_total.index, df_total[column], label=column)


plt.title("Crecimiento de la poblaciÃ³n (1960-2023)", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Poblation", fontsize=12)
plt.legend(title="Country")
plt.grid(True, linestyle="--", alpha=0.7)

plt.xticks(df_total.index[::10], rotation=45)
```
