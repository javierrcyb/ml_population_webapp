```python
import pandas as pd
import matplotlib.pyplot as plt

df_male = pd.read_csv("data\population_male.csv", quotechar='"')
df_total = pd.read_csv("data\population_total.csv", quotechar='"')
```

```python
df_diferencia = df_male[~df_male['Country Name'].isin(df_total['Country Name'])]
df_diferencia
```

```python
df_male = df_male.drop(columns=["Country Code", "Indicator Name", "Indicator Code", "Unnamed: 68"])
df_male.set_index('Country Name', inplace=True)
df_male = df_male.T
df_male.index.name = "Years"
df_male.head()
```

```python
df_male.isnull().sum().sum()
```

```python
df_male = df_male.drop(columns=["World"])
```

```python
plt.figure(figsize=(10, 5))
for column in df_male.columns[:10]:
    plt.plot(df_male.index, df_male[column], label=column)


plt.title("Crecimiento de la población (1960-2023)", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Poblation", fontsize=12)
plt.legend(title="Country")
plt.grid(True, linestyle="--", alpha=0.7)

plt.xticks(df_male.index[::10], rotation=45)
```
